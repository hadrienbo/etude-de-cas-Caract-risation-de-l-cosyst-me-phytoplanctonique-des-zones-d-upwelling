```python
def replacenans(A,M):
  s=0
  B=np.zeros((A.shape[0],A.shape[1],A.shape[2]))
  for i in range(A.shape[0]):
    print(i/A.shape[0]*100,'%')
    for j in range(A.shape[1]):
      for k in range(A.shape[2]):
        if np.isnan(A[i,j,k]):
          B[i,j,k]=M[j,k]
        else:
          B[i,j,k]=A[i,j,k]
  return B
```
